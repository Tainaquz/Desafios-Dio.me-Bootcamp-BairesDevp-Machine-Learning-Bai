```python
import cv2
import matplotlib.pyplot as plt
def exibir (image):
  imagem = cv2.imread(image)
  figura = plt.gcf() # pega a figura da pasta gcf e plota no matplot
  figura.set_size_inches(10,8) #tamanho geradi na saida
  plt.axis("off")
  plt.imshow(cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB))
  plt.show()
```

```python
exibir('predictions.jpg')
```
